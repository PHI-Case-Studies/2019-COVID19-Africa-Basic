# First We Process the Employment Data(Ratios)

```python
# import pandas 
import pandas as pd
#read the employment data
empdf1 = pd.read_csv('data/API_SL.EMP.TOTL.SP.MA.ZS_DS2_en_csv_v2_1070219.csv', skiprows = 3)
empdf1
```

# Now we will process the data for Confirmed Covid Cases in Africa

```python
worldcovid_df = pd.read_csv('data/time_series_covid19_deaths_global_narrow.csv')
worldcovid_df
```

```python
# sort values by Date
worldcovid_df1 = worldcovid_df.sort_values(by="Date")
worldcovid_df1
```

```python
worldcovid_df2 = worldcovid_df1.drop([0])
worldcovid_df2
```

```python
#drop Province/State Columns
worldcovid_df2.drop(['Province/State'], axis=1, inplace=True)
```

```python
worldcovid_df2
```

```python
worldcovid_df2.columns = ['CNTRY_NAME', 'Lat', 'Long', 'Date', 'Confirmed_Cases']
```

```python
worldcovid_df2.CNTRY_NAME.unique()
```

```python
worldcovid_df2.loc[worldcovid_df2['CNTRY_NAME']=='Congo (Kinshasa)', 'CNTRY_NAME'] = 'Dem Rep Congo'
worldcovid_df2.loc[worldcovid_df2['CNTRY_NAME']=='Congo (Brazzaville)', 'CNTRY_NAME'] = 'Rep Congo'
```

```python
worldcovid_df2.to_pickle('data/worldcovid_df2_pickle.pkl')
```
