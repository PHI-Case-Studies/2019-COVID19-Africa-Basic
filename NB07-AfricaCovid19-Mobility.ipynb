# This notebook will examine trends in Mobility in Africa following the Covid-19 Outbreak
> * Lockdown( travel restrictions&shutdowns) was one of the key interventions implentented by most african countries to contain the spread of Covid-19
> * However some countries eg Tanzania did not implement lockdowns
> * Mobility can also be used as a proxy for Social Distancing
> * We will use line plots and heatmaps to visualize our mobility and cases data

```python
import pandas as pd
```

```python
# read the mobility data csv file
mobdf = pd.read_csv('data/Global_Mobility_Report.csv')
mobdf
```

```python
#rename the column headings
mobdf.columns = ['Country Code', 'CNTRY_NAME', 'region1', 'region2', 'date', 'retail_and_recreation', 'grocery_and_pharmacy', 'parks', 'transit_stations', 'workplaces', 'residential']
mobdf
```

```python
#drop columns region1  and region2 
mobdf.drop(['region1', 'region2'],axis=1, inplace=True)
mobdf
```

```python
mobdf
```

```python
# Select df.desired column row == Zimbabwe
zimmobility = mobdf[(mobdf.Country_Region == "Zimbabwe")]
zimmobility
```

```python
# Using plotly.express
import plotly.express as px

import pandas as pd


fig = px.line(zimmobility, x='date', y='retail_and_recreation', range_x=['2020-02-15','2020-05-18'], title='Zimabwe retail and recreation mobility trends',)

fig.show()
```

```python
worldcovid_df2= pd.read_pickle('data/worldcovid_df2_pickle.pkl')
worldcovid_df2
```

```python
worldcovid_df3=worldcovid_df2.columns = ['CNTRY_NAME', 'Lat', 'Long', 'Date', 'Confirmed_Cases']
```

```python
# Select df.desired column row == Zimbabwe
zimcovidf = worldcovid_df2[(worldcovid_df2.CNTRY_NAME == "Zimbabwe")]
zimcovidf
```

```python
# Using plotly.express
import plotly.express as px

import pandas as pd

fig = px.line(zimcovidf, x='Date', y='Confirmed_Cases', range_x=['2020-02-15','2020-05-18'], title='Zimabwe Covid-19 Cases Trends',)

fig.show()
```

```python
africapopdf1= pd.read_pickle('data/africapopdf1_pickle.pkl')
africapopdf1
```

```python
africamobility = mobdf.merge(africapopdf1, on='CNTRY_NAME')
africamobility
```

```python
import plotly.graph_objects as go
import datetime
import numpy as np
np.random.seed(1)

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
min_date=africamobility['date'].min()
max_date=africamobility['date'].max()
```

```python
min_date
```

```python
fig = go.Figure(data=go.Heatmap(
        z=africamobility['retail_and_recreation'],
        x=africamobility['date'],
        y=africamobility['CNTRY_NAME'],       
        colorscale='plasma',
        colorbar=dict(thickness=20,lenmode='fraction',len=0.8,
                      tickcolor='black',tickfont=dict(size=10),y=0.8,
                     title='mobility retail&recreation')
))

fig.update_layout(
    title='Heat Map, Mobility in Africa, '+min_date+' to '+max_date, 
    titlefont=dict(size=14),
    yaxis_nticks=len(africamobility['CNTRY_NAME'].unique()),
    xaxis_nticks=len(africamobility['date'].unique()),
    width=1100,
    height=700,
    yaxis=dict(title_text="Country", titlefont=dict(size=12),tickfont=dict(size=9)),
    xaxis=dict(title_text="Report Date",titlefont=dict(size=12),tickfont=dict(size=9)),
    margin=dict(l=50, r=50, b=100, t=100,pad=4)
)

fig.show()
```

```python
fig = go.Figure(data=go.Heatmap(
        z=africamobility['residential'],
        x=africamobility['date'],
        y=africamobility['CNTRY_NAME'],       
        colorscale='plasma',
        colorbar=dict(thickness=20,lenmode='fraction',len=0.4,
                      tickcolor='black',tickfont=dict(size=10),y=0.8,
                     title='mobility residential')
))

fig.update_layout(
    title='Heat Map, Mobility in Africa, '+min_date+' to '+max_date, 
    titlefont=dict(size=14),
    yaxis_nticks=len(africamobility['CNTRY_NAME'].unique()),
    xaxis_nticks=len(africamobility['date'].unique()),
    width=1000,
    height=700,
    yaxis=dict(title_text="Country", titlefont=dict(size=12),tickfont=dict(size=9)),
    xaxis=dict(title_text="Report Date",titlefont=dict(size=12),tickfont=dict(size=9)),
    margin=dict(l=50, r=50, b=100, t=100,pad=4)
)

fig.show()
```
